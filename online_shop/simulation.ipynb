```python
from faker import Faker # data
import random
import pandas as pd
import faker_commerce

fake = Faker("de_DE")
fake.add_provider(faker_commerce.Provider)

rows = []
n_user = 10_000

for _ in range(n_user):
    first_name = fake.first_name()
    last_name = fake.last_name()
    user_id = fake.unique.uuid4()
    user = {
        "id" : user_id,
        "name": first_name + " " + last_name,
    }
    rows.append(user)

df = pd.DataFrame(rows)
df
```

```python
rows = []
n_products = 500 

for _ in range(n_products):
    product = {
        "id": fake.unique.uuid4(),
        "name": fake.ecommerce_name(),
        "price": float(random.randint(10,1000)) + 0.99,
        "available": random.randint(0,40),
    }
    rows.append(product)

df_product = pd.DataFrame(rows)
df_product
```

```python
df_product_m = df_product.copy() # in pandas this is deep copy 

# cat 1 -> costs more than 600
# cat 2 -> costs 300 - 600
# cat 3 -> costs 200 - 300
# cat 4 -> costs <200


def price_category(price):
    if price>600:
        return "Category 1"
    elif price > 300:
        return "Category 2"
    elif price > 200:
        return "Category 3"
    else:
        return "Category 4"

df_product_m["price_category"]=df_product_m["price"].apply(lambda p: price_category(p)) # Lambda notation
df_product_m
```

```python
# TODO
# create a new column which indicates whether the available quanitity of a product is equal to zero

df_product_m["out_of_stock"] = df_product_m["available"]==0
df_product_m


def out_of_stock(quantity):
    if quantity == 0:
        return True
    else:
        return False

df_product_m["out_of_stock"] =df_product_m["available"].apply(lambda q: out_of_stock(q))
sum(df_product_m["out_of_stock"]) # count
```

```python
def filter_category_one(df): return df[df["price_category"] == "Category 1"]
def filter_on_stock(df): return df[df["out_of_stock"] == False]

# + we sort the rows alongside the price (descending)

df_category_one = (df_product_m
    .pipe(filter_category_one)
    .sort_values("price", ascending = False)
)
#df_category_one

df_category_one_available = (df_product_m
    .pipe(filter_category_one)
    .pipe(filter_on_stock)
    .sort_values("price", ascending = False)
)
df_category_one_available
```

```python
agg_df = df_product_m.groupby(["price_category"])[["price","available"]].agg(
    [
        "mean",
        "std",
        "median",
        "min",
        "max"
    ]
)
agg_df
```
